```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
F = float(input('fc = '))
E = float(input('ec^2/2 = '))
S = float(input('sigma crit = '))
precisao = int(input('numero de pontos = '))
xmax = S+1
X = np.arange(0,xmax,xmax/precisao)
Fx = np.zeros(precisao)
for i in range(1,precisao-1):
    Fx[i] = F/X[i] - (1+E*1/cos(pi/2*(X[i]/S)**(1/2)))
for i in range(1,precisao-2):
    if(Fx[i]*Fx[i+1] <= 0):
        valor = X[i]
        print('Ïƒcrit = ', X[i])
        break
zerro = np.zeros(precisao)
plt.subplots(figsize=(10,10)) 
plt.xlim(valor-(xmax/10),valor+(xmax/10))
plt.ylim(-10,10)
plt.plot(X, zerro)
txt = 'sigma crit = '+str(np.round(valor,3))
plt.plot(X, Fx, label = txt)
plt.legend()
plt.show
```
