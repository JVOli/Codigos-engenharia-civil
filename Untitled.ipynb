```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
from math import *
abacate = (4-(2)**(1/2))/14                                                     #medida base triangulo 45°
#Posição dos nós
nos = np.array([[0,0],                                                          #nó A (0)
                [1,0],                                                          #nó B (1)
                [0.5,0],                                                        #nó C (2)
                [abacate,0],                                                    #nó 3
                [2*abacate,0],                                                  #nó 4
                [1-2*abacate,0],                                                #nó 5
                [1-abacate,0],                                                  #nó 6
                [abacate/2,abacate/2],                                          #nó 7
                [3*abacate/2, abacate/2],                                       #nó 8
                [1-3*abacate/2,abacate/2],                                      #nó 9
                [1-abacate/2,abacate/2],                                        #nó 10
                [abacate,abacate],                                              #nó 11
                [2*abacate,abacate*1.2],                                        #nó 12
                [0.5, abacate*1.3],                                             #nó 13
                [1-2*abacate,abacate*1.2],                                      #nó 14
                [1-abacate,abacate]])                                           #nó 15
#conectividade das barras em coordenadas
barras_coord = np.array([[nos[0,:],nos[3,:]],                                   #barra 0
                   [nos[0,:],nos[7,:]],                                         #barra 1
                   [nos[3,:],nos[7,:]],                                         #barra 2
                   [nos[3,:],nos[11,:]],                                        #barra 3
                   [nos[3,:],nos[8,:]],                                         #barra 4
                   [nos[3,:],nos[4,:]],                                         #barra 5
                   [nos[8,:],nos[11,:]],                                        #barra 6
                   [nos[8,:],nos[4,:]],                                         #barra 7
                   [nos[7,:],nos[11,:]],                                        #barra 8
                   [nos[11,:],nos[12,:]],                                       #barra 9
                   [nos[12,:],nos[13,:]],                                       #barra 10
                   [nos[12,:],nos[4,:]],                                        #barra 11
                   [nos[13,:],nos[4,:]],                                        #barra 12
                   [nos[13,:],nos[2,:]],                                        #barra 13
                   [nos[13,:],nos[5,:]],                                        #barra 14
                   [nos[13,:],nos[14,:]],                                       #barra 15
                   [nos[2,:],nos[4,:]],                                         #barra 16
                   [nos[2,:],nos[5,:]],                                         #barra 17
                   [nos[14,:],nos[5,:]],                                        #barra 18
                   [nos[14,:],nos[15,:]],                                       #barra 19
                   [nos[5,:],nos[9,:]],                                         #barra 20
                   [nos[5,:],nos[6,:]],                                         #barra 21
                   [nos[6,:],nos[9,:]],                                         #barra 22
                   [nos[6,:],nos[15,:]],                                        #barra 23
                   [nos[6,:],nos[10,:]],                                        #barra 24
                   [nos[9,:],nos[15,:]],                                        #barra 25
                   [nos[15,:],nos[10,:]],                                       #barra 26
                   [nos[1,:],nos[10,:]],                                        #barra 27
                   [nos[1,:],nos[6,:]]])                                        #barra 28
#conectividade das barras em inteiros
barras = np.array([[0,3],                                                       #barra 0
                   [0,7],                                                       #barra 1
                   [3,7],                                                       #barra 2
                   [3,11],                                                      #barra 3
                   [3,8],                                                       #barra 4
                   [3,4],                                                       #barra 5
                   [8,11],                                                      #barra 6
                   [8,4],                                                       #barra 7
                   [7,11],                                                      #barra 8
                   [11,12],                                                     #barra 9
                   [12,13],                                                     #barra 10
                   [12,4],                                                      #barra 11
                   [13,4],                                                      #barra 12
                   [13,2],                                                      #barra 13
                   [13,5],                                                      #barra 14
                   [13,14],                                                     #barra 15
                   [2,4],                                                       #barra 16
                   [2,5],                                                       #barra 17
                   [14,5],                                                      #barra 18
                   [14,15],                                                     #barra 19
                   [5,9],                                                       #barra 20
                   [5,6],                                                       #barra 21
                   [6,9],                                                       #barra 22
                   [6,15],                                                      #barra 23
                   [6,10],                                                      #barra 24
                   [9,15],                                                      #barra 25
                   [15,10],                                                     #barra 26
                   [1,10],                                                      #barra 27
                   [1,6]])                                                      #barra 28
#areas barras
area_barra = np.array([[1.E-4],                                                 #área da barra 0
                       [1.E-4],                                                 #área da barra 1
                       [1.E-4],                                                 #área da barra 2
                       [1.E-4],                                                 #área da barra 3
                       [1.E-4],                                                 #área da barra 4
                       [1.E-4],                                                 #área da barra 5
                       [1.E-4],                                                 #área da barra 6
                       [1.E-4],                                                 #área da barra 7
                       [1.E-4],                                                 #área da barra 8
                       [1.E-4],                                                 #área da barra 9
                       [1.E-4],                                                 #área da barra 10
                       [1.E-4],                                                 #área da barra 11
                       [1.E-4],                                                 #área da barra 12
                       [1.E-4],                                                 #área da barra 13
                       [1.E-4],                                                 #área da barra 14
                       [1.E-4],                                                 #área da barra 15
                       [1.E-4],                                                 #área da barra 16
                       [1.E-4],                                                 #área da barra 17
                       [1.E-4],                                                 #área da barra 18
                       [1.E-4],                                                 #área da barra 19
                       [1.E-4],                                                 #área da barra 20
                       [1.E-4],                                                 #área da barra 21
                       [1.E-4],                                                 #área da barra 22
                       [1.E-4],                                                 #área da barra 23
                       [1.E-4],                                                 #área da barra 24
                       [1.E-4],                                                 #área da barra 25
                       [1.E-4],                                                 #área da barra 26
                       [1.E-4],                                                 #área da barra 27
                       [1.E-4]])                                                #área da barra 28
#Numero dos nos A B
Apoio_A = 0
Apoio_B = 1      
#Numero do no C Carga C
Apoio_C = 2
#Carga Maxima = 167.6
Carga_C = 200  
#Calcular peso da estrutra
def peso_estrutura(Nos, Barras, Areas):
  Densidade = 1.547*10**3
  Peso = 0                                                                      #Variavel que vai somar os pesos
  for i in range(0,np.shape(barras[:,0])[0]):                                   #Somar o peso de todas as barras
    Peso = Densidade * area_barra[i,0] * ((nos[barras[i,0],0]-nos[barras[i,1],0])**2+(nos[barras[i,0],1]-nos[barras[i,1],1])**2)**(1/2) + Peso
  return Peso
Peso = peso_estrutura(nos, barras, area_barra)
num_nos = np.shape(nos)[0]                                                      #Numero Nos                     
num_barras = np.shape(barras)[0]                                                #Numero Barras
E = 3.6*10**8                                                                   #Constante E
def solucao_estrutural(nodes,barras,area_bar,E,nnodes,nbarras,
					   no_apoio_A,no_apoio_B,no_carga,carga_aplicada):

	# ________________________________________
	#
	# ALGORITMO DE SOLUÇÃO ESTRUTURAL DE
	# TRELIÇA PLANA LINEAR
	# ________________________________________
	#
	# > nodes          : coordenadas x e y de cada nó
	# > barras         : nó inicial e nó final de cada barra
	# > area_bar       : a área da seção transversal de cada barra
	# > E              : módulo de elasticidade do material
	# > nnodes         : número de nós
	# > nbarras        : número de barras
	# > no_apoio_A     : apoio 1 (ux = uy = 0.0)
	# > no_apoio_B     : apoio 2 (ux = uy = 0.0)
	# > no_carga       : nó onde a carga F = (0.0,carga_aplicada) é aplicada
	# > carga_aplicada : valor da carga vertical aplicada 
	# < fbarra         : força na direção de cada barra ( tração = (+), compressão = (-) )
	# ________________________________________

#copia dados da geometria da estrutura	
	nodesc = np.copy(nodes)
	barrasc = np.copy(barras)
	area_barc = np.copy(area_bar)
#inicializa arranjos
	l_bar = np.zeros( barrasc.shape[0] )
	cθ_bar = np.zeros( barrasc.shape[0] )
	sθ_bar = np.zeros( barrasc.shape[0] )
	kl = np.zeros( (4,4,nbarras) )
	Kg = np.zeros( (2*nnodes,2*nnodes) )
	Fg = np.zeros( (2*nnodes,1) )
	cc = np.zeros( (2*nnodes,1),dtype='int32' )
	fbarra = np.zeros (nbarras)
#aplica a carga (carga_aplicada) na direção y negativa do nó no_carga
	Fg[2*no_carga+1,0] = -carga_aplicada
#aplica condição de contorno de apoio (deslocamentos nulos) no nós no_apoio_A e no_apoio_B
	cc[2*no_apoio_A,0]   = 1
	cc[2*no_apoio_A+1,0] = 1
	cc[2*no_apoio_B,0]   = 1
	cc[2*no_apoio_B+1,0] = 1
#calcula comprimendo e ângulos das barras
	for i in range(0,nbarras):
		no1 = barrasc[i,0]
		no2 = barrasc[i,1]
		dx = nodesc[no1,0]-nodesc[no2,0]
		dy = nodesc[no1,1]-nodesc[no2,1]
		l_bar[i] = sqrt(dx**2 + dy**2)
		cθ_bar[i] = -dx/l_bar[i]
		sθ_bar[i] = -dy/l_bar[i]
#calcula rigidez kl de cada barra:
	for i in range(0,nbarras):
		kb = ( E * area_barc[i] / l_bar[i] )
		kl[0,0,i] =  kb * ( cθ_bar[i] )**2
		kl[0,1,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[0,2,i] = -kb * ( cθ_bar[i] )**2
		kl[0,3,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[1,0,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[1,1,i] =  kb * ( sθ_bar[i] )**2
		kl[1,2,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[1,3,i] = -kb * ( sθ_bar[i] )**2
		kl[2,0,i] = -kb * ( cθ_bar[i] )**2
		kl[2,1,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[2,2,i] =  kb * ( cθ_bar[i] )**2
		kl[2,3,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[3,0,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[3,1,i] = -kb * ( sθ_bar[i] )**2
		kl[3,2,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[3,3,i] =  kb * ( sθ_bar[i] )**2
#montagem da rigidez Kg do sistema linear:
	for i in range(0,nbarras):
		no1 = barrasc[i,0]
		no2 = barrasc[i,1]
		noAx = no1*2
		noAy = noAx + 1
		noBx = no2*2
		noBy = noBx + 1
		Kg[noAx,noAx] = Kg[noAx,noAx] + kl[0,0,i]
		Kg[noAx,noAy] = Kg[noAx,noAy] + kl[0,1,i]
		Kg[noAy,noAx] = Kg[noAy,noAx] + kl[1,0,i]
		Kg[noAy,noAy] = Kg[noAy,noAy] + kl[1,1,i]
		Kg[noAx,noBx] = Kg[noAx,noBx] + kl[0,2,i]
		Kg[noAx,noBy] = Kg[noAx,noBy] + kl[0,3,i]
		Kg[noAy,noBx] = Kg[noAy,noBx] + kl[1,2,i]
		Kg[noAy,noBy] = Kg[noAy,noBy] + kl[1,3,i]
		Kg[noBx,noAx] = Kg[noBx,noAx] + kl[2,0,i]
		Kg[noBy,noAx] = Kg[noBy,noAx] + kl[2,1,i]
		Kg[noBx,noAy] = Kg[noBx,noAy] + kl[3,0,i]
		Kg[noBy,noAy] = Kg[noBy,noAy] + kl[3,1,i]
		Kg[noBx,noBx] = Kg[noBx,noBx] + kl[2,2,i]
		Kg[noBx,noBy] = Kg[noBx,noBy] + kl[2,3,i]
		Kg[noBy,noBx] = Kg[noBy,noBx] + kl[3,2,i]
		Kg[noBy,noBy] = Kg[noBy,noBy] + kl[3,3,i]
#aplica condições de contorno sobre a rigidez Kg do sistema linear	
	for i in range(0,2*nnodes):
		if ( cc[i] == 1 ):
			Kg[:,i] = 0.0
			Kg[i,:] = 0.0
			Kg[i,i] = 1.0
#calcula os deslocamento x e y de cada nó do sistema linear Ax=b,
#onde A=Kg, x=uxy, b=Fg, empregando eliminação de Gauss
	uxy = np.linalg.solve(Kg,Fg)
#calcula força interna na direção de cada barra
	for i in range(0,nbarras):
		no1 = barrasc[i,0]
		no2 = barrasc[i,1]
		ul1x = uxy[no1*2]
		ul1y = uxy[no1*2+1]
		ul2x = uxy[no2*2]
		ul2y = uxy[no2*2+1]
		fbarra[i] = ( ( cθ_bar[i]*kl[2,0,i] + sθ_bar[i]*kl[3,0,i] ) * ul1x + 
    				  ( cθ_bar[i]*kl[2,1,i] + sθ_bar[i]*kl[3,1,i] ) * ul1y +
					  ( cθ_bar[i]*kl[2,2,i] + sθ_bar[i]*kl[3,2,i] ) * ul2x +
					  ( cθ_bar[i]*kl[2,3,i] + sθ_bar[i]*kl[3,3,i] ) * ul2y )
	E = 3.6*10**8
	Rimin = 10000000                                                              #Valor alto onde sera comparado com os Ri de todas as barras
	for i in range(0, nbarras):                                                   #Calcular Ri de todas as barras
#Equacoes
		if(fbarra[i] >=0.0001):
			Ri = 1.676*10**6-(fbarra[i]/area_bar[i,0])
		if(fbarra[i] <0.0001):
			Termomin = (np.pi*E*area_bar[i,0])/(4*((l_bar[i])**(1/2))**2)
		if(Termomin < 1.676*10**6):
			Ri = Termomin - (fbarra[i]/area_bar[i,0])
		else:
				Ri = 1.676*10**6 - (fbarra[i]/area_bar[i,0])               
#Salva o menor
		if(Ri<Rimin):                                                               #Compara o menor Ri com o atual                                                            
			Rimin = Ri
#saída de valores 
	return fbarra, Rimin                                                          #Retorna as forcas das barras[0:28], CRi       
forca_barras, CRi = solucao_estrutural(nos,barras,area_barra,E,num_nos,num_barras,
				             	  Apoio_A,Apoio_B,Apoio_C,Carga_C)
ForcaOriginal = np.copy(forca_barras)                                           #Deixar uma versão sem modificacoes salva das forcas para a otimizacao
for i in range(0,np.shape(forca_barras)[0]):                                    #Printar as forcas
	print('Força na barra entre os nos '+ str(barras[i,0])+'-'+str(barras[i,1])+' / Barra '+ str(i)+' = {:^+15.6E}'.format(float(round(forca_barras[i],5))) )
#Plota a estrutura
def plotar_estrutura(Nos, Barras, N_apoio_A, N_apoio_B, N_apoio_C, Pes, Carga):
  plt.figure(1, figsize=(8,8))
  plt.grid(True)
  plt.title('PESO = '+str(round(Pes, 5))+' Kg')
  plt.xlabel('x', fontsize = 15)
  plt.ylabel('y', fontsize = 15)
  plt.axis([-0.2, +1.2, -0.2, np.max(nos[:,1])+0.2])                            #Defini eixos com margem 0.2     
  for i in range(0,np.shape(Barras[:,0,0])[0]):                                 #Plota todas as barras
    plt.plot(Barras[i,:,0], Barras[i,:,1], c='b')
  plt.scatter(Nos[:,0], Nos[:,1], c='r')                                        #Plota todos os nos
#Plota os apoios AB e seta 
  plt.plot(Nos[N_apoio_B,0],Nos[N_apoio_B,1]-0.01,'^',c='k',ms=15.0)
  plt.plot(Nos[N_apoio_A,0],Nos[N_apoio_A,1]-0.01,'^',c='k',ms=15.0)
  plt.arrow(Nos[N_apoio_C,0],Nos[N_apoio_C,1],0, -0.030, head_width=0.02, head_length=0.020, fc='k', ec='k')
  plt.show()
plotar_estrutura(nos, barras_coord, Apoio_A, Apoio_B, Apoio_C, Peso, Carga_C)
#Iteracao
def f(x):
#Mesma teoria da linha 250                                                                       
  forca_barras, CRi = solucao_estrutural(nos,barras,area_barra,E,num_nos,num_barras, Apoio_A,Apoio_B,Apoio_C,x)
  Rimin = 100000000                                                             
  for i in range(0, num_barras):
    if(forca_barras[i] >=0.001):
      Ri = 1.676*10**6-(forca_barras[i]/area_barra[i,0])
    if(forca_barras[i] <0.001):
      Termomin = (np.pi*E*area_barra[i,0])/(4*(((barras_coord[i,0,0]-barras_coord[i,1,0])**2+(barras_coord[i,0,1]-barras_coord[i,1,1])**2)**(1/2))**2)
      if(Termomin < 1.676*10**6):
        Ri = Termomin - (forca_barras[i]/area_barra[i,0])
      if(Termomin>=1.676*10**6):
        Ri = 1.676*10**6 - (forca_barras[i]/area_barras[i,0])
    if(Ri<Rimin):
      Rimin = Ri
  return Rimin                                                                  
#Encontrar a carga limite
#Metodo de iteracao
def iterpontofixo(x0,es,imax):                                                  #Recebe parametros:(Carga inicial, Erro aceitavel, Maximo de iteracoes)
  for i in range(0,imax+1):                                                     #Roda ate o maximo de iteracoes           
    if(x0>50):                                                                  #Varia o metodo de acordo com o valor visando a eficiencia
      if ( i == 0 ):
        Rimin1 = f(x0)                                                          
        xi = x0 + Rimin1*x0/10/(1.676*10**6)
      else:
        Rimin1 = f(xi)
        xi = xi +  Rimin1*x0/10/(1.676*10**6)
      er = ((Rimin1)**2)**(1/2)
      if ( er < es ):                                                           #Para se o erro esta aceitavel  
        break
    if(x0<50):                                                                  #Varia o metodo de acordo com o valor visando a eficiencia
      if ( i == 0 ):
        Rimin1 = f(x0)
        xi = x0 + Rimin1*(x0+5)/(1.676*10**6) 
      else:
        Rimin1 = f(xi)
        xi = xi +  Rimin1*(x0+5)/(1.676*10**6)
      er = ((Rimin1)**2)**(1/2)
      if ( er < es ):                                                           #Para se o erro esta aceitavel  
        break
  return xi                                                                     #Retorna Carga de Colapso
print('Carga de colapso =', round(iterpontofixo(Carga_C,0.001,200), 5), 'kg')   #Printa carga de colapso sem otimizacao
print('-----------------------------------------------------------------------------------------------')
print('Otimizacao')
print()
#Metodo de otimizacao
Condicoes = True                                                                
while(Condicoes == True):                                                       #Testa se tem mudancas significantes
  contador = 0                                                                  #Conta o numero significantes, se for igual ao numero ciclos: para
  for repeticao in range(0, 2):                                                 #Realiza 2 cilcos antes de zerar o contador
    CRiold13 = f(Carga_C)                                                       #Antigo Ri da Carga 
    for t in range(0,10000):                                                          #Otimiza 100 posicao y barra 12/14
      Teste = True
      if(Teste == True):
        Pesoinicial = peso_estrutura(nos, barras, area_barra)
        delta = 1                                                               #Delta em porcentagem do peso que sera alterado
        nos[12,1] = nos[12,1] - nos[12,1]*delta/100                             #Novo comprimento da barra 12
        nos[14,1] = nos[14,1] - nos[14,1]*delta/100                             #Novo comprimento da barra 14
        CRinew13 = f(Carga_C)                                                   #Novo CRi
        pg = (Pesoinicial - peso_estrutura(nos, barras, area_barra) ) / (nos[14,1]*delta/100)      #Gradiente do Peso por diferencas finitas   
        fg = (CRiold13 - CRinew13 ) / (nos[14,1]*delta/100)                     #Gradiente do CRi por diferencas finitas
        CRiold13 = CRinew13                                                     #Transforma o novo comprimento no antigo para executar a otimizacao novamente
        if(fg > 0 or pg < 0):                                                   #Para de rodar se for a otimizacao maxima
          if(t<1):                                                              #Testa mudanca significativa por loop  
            contador = 1 + contador
          break 
    for t in range(0,10000):                                                    #Otimiza 100 posicao y da barra 13
      if(Teste == True):
        Pesoinicial = peso_estrutura(nos, barras, area_barra)
        delta = 1                                                               #Delta em porcentagem do peso que sera alterado
        nos[13,1] = nos[13,1] - nos[13,1]*delta/100                             #Novo comprimento da barra 13   
        CRinew13 = f(Carga_C)                                                   #Novo CRi
        pg = (Pesoinicial - peso_estrutura(nos, barras, area_barra) ) / (nos[13,1]*delta/100)     #Gradiente do Peso por diferencas finitas   
        fg = (CRiold13 - CRinew13 ) / (nos[13,1]*delta/100)                     #Gradiente do CRi por diferencas finitas
        CRiold13 = CRinew13                                                     #Transforma o novo comprimento no antigo para executar a otimizacao novamente
        if(fg > 0.0001 or pg < 0):                                              #Para de rodar se for a otimizacao maxima
          if(t<1):                                                              #Testa mudanca significativa por loop  
            contador = 1 + contador
          break

  if(contador == 4):                                                            #Testa o contador                                                      
      Condicoes = False

print('Otimizacao comprimento Barra 14 concluida')
print('Novo valor do comprimento 14: ', nos[14,1],' Metros')
print()
print('Otimizacao comprimento Barra 12 concluida')
print('Novo valor do comprimento 12: ', nos[12,1],' Metros')
print()
print('Otimizacao comprimento Barra 13 concluida')
print('Novo valor do comprimento 13: ', nos[13,1],' Metros')
print()
Condicoes = True                                                                
while(Condicoes == True):                                                       #Testa se tem mudancas significantes
  contador = 0                                                                  #Conta o numero significantes, se for igual ao numero ciclos: para
  for repeticao in range(0, 2):                                                 #Realiza 2 cilcos antes de zerar o contador
    CRiold13 = f(Carga_C)  
    for t in range(0,1000):                                                     #Otimiza 100 area da barra 16/17
      if(Condicoes == True):
        delta = 2                                                              #Delta em porcentagem do peso que sera alterado
        area_barra[16] = area_barra[16] + area_barra[16]*delta/100              #Nova area da barra 16 
        area_barra[17] = area_barra[17] + area_barra[17]*delta/100              #Nova area da barra 17     
        CRinew13 = f(Carga_C)                                                   #Novo CRi
        fg = (CRinew13 - CRiold13 ) / (area_barra[16]*delta/100)                #Gradiente CRi por diferencas finitas
        CRiold13 = CRinew13                                                     #Transforma a nova area 13 na antiga para executar a otimizacao novamente
        if(fg <= 0 or peso_estrutura(nos, barras, area_barra)>= 0.75):       #Para de rodar se for a otimizacao maxima ou atingir um dos limites'
          if(peso_estrutura(nos, barras, area_barra)>= 0.75):                   #Testa condicao de peso
            area_barra[16] = area_barra[16] - area_barra[16]*2/100              #Como nao houve mudanca significativa, retorna ao estado anterior 
            area_barra[17] = area_barra[17] - area_barra[17]*2/100            
          if(t<1):                                                              #Testa mudanca significativa por loop  
            contador = 1 + contador
          break                                            
    for t in range(0,1000):                                                     #Otimiza 100 area da barra 13
      if(Condicoes == True):
        delta = 2                                                               #Delta em porcentagem do peso que sera alterado
        area_barra[13] = area_barra[13] + area_barra[13]*delta/100              #Nova area da barra 13     
        CRinew13 = f(Carga_C)                                                   #Novo CRi
        fg = (CRinew13 - CRiold13 ) / (area_barra[13]*delta/100)                #Gradiente CRi por diferencas finitas
        CRiold13 = CRinew13                                                     #Transforma a nova area na antiga para executar a otimizacao da area novamente
        if(fg <= 0 or peso_estrutura(nos, barras, area_barra)>= 0.75):       #Para de rodar se for a otimizacao maxima ou atingir um dos limites
          if(peso_estrutura(nos, barras, area_barra)>= 0.75):                   #Testa condicao de peso
            area_barra[13] = area_barra[13] - area_barra[13]*2/100              #Como nao houve mudanca significativa, retorna ao estado anterior 
          if(t<1):                                                              #Testa mudanca significativa por loop 
            contador = 1 + contador 
          break
        CRiold13 = f(Carga_C)  
    for t in range(0,1000):                                                     #Otimiza 100 area da barra 16/17
      if(Condicoes == True):
        delta = 2                                                               #Delta em porcentagem do peso que sera alterado
        area_barra[11] = area_barra[11] + area_barra[11]*delta/100              #Nova area da barra 16 
        area_barra[18] = area_barra[18] + area_barra[18]*delta/100              #Nova area da barra 17     
        CRinew13 = f(Carga_C)                                                   #Novo CRi
        fg = (CRinew13 - CRiold13 ) / (area_barra[16]*delta/100)                #Gradiente CRi por diferencas finitas
        CRiold13 = CRinew13                                                     #Transforma a nova area 13 na antiga para executar a otimizacao novamente
        if(fg <= 0 or peso_estrutura(nos, barras, area_barra)>= 0.75):       #Para de rodar se for a otimizacao maxima ou atingir um dos limites'
          if(peso_estrutura(nos, barras, area_barra)>= 0.75):                   #Testa condicao de peso
            area_barra[11] = area_barra[11] - area_barra[11]*2/100              #Como nao houve mudanca significativa, retorna ao estado anterior 
            area_barra[18] = area_barra[18] - area_barra[18]*2/100            
          if(t<1):                                                              #Testa mudanca significativa por loop  
            contador = 1 + contador
          break   

    if(contador == 6):                                                          #Testa o contador                                                      
      Condicoes = False                                                         #Termina o while
print('Otimizacao da area da Barra 13 concluida')
print('Novo valor do comprimento 13: ', area_barra[13],' Metros')
print()
print('Otimizacao da area da Barra 16 concluida')
print('Novo valor do comprimento 16: ', area_barra[16],' Metros')
print()
print('Otimizacao da area da Barra 17 concluida')
print('Novo valor do comprimento 17: ', area_barra[17],' Metros')
print()
print('Otimizacao da area da Barra 11 concluida')
print('Novo valor do comprimento 11: ', area_barra[11],' Metros')
print()
print('Otimizacao da area da Barra 18 concluida')
print('Novo valor do comprimento 18: ', area_barra[18],' Metros')
print()
print('-----------------------------------------------------------------------------------------------')
print('Nova carga de colapso =', round(iterpontofixo(Carga_C,0.001,200), 5), 'kg')
#Reinicia barras_coord com os novos valores para a plotagem
barras_coord = np.array([[nos[0,:],nos[3,:]],                                   #barra 0
                   [nos[0,:],nos[7,:]],                                         #barra 1
                   [nos[3,:],nos[7,:]],                                         #barra 2
                   [nos[3,:],nos[11,:]],                                        #barra 3
                   [nos[3,:],nos[8,:]],                                         #barra 4
                   [nos[3,:],nos[4,:]],                                         #barra 5
                   [nos[8,:],nos[11,:]],                                        #barra 6
                   [nos[8,:],nos[4,:]],                                         #barra 7
                   [nos[7,:],nos[11,:]],                                        #barra 8
                   [nos[11,:],nos[12,:]],                                       #barra 9
                   [nos[12,:],nos[13,:]],                                       #barra 10
                   [nos[12,:],nos[4,:]],                                        #barra 11
                   [nos[13,:],nos[4,:]],                                        #barra 12
                   [nos[13,:],nos[2,:]],                                        #barra 13
                   [nos[13,:],nos[5,:]],                                        #barra 14
                   [nos[13,:],nos[14,:]],                                       #barra 15
                   [nos[2,:],nos[4,:]],                                         #barra 16
                   [nos[2,:],nos[5,:]],                                         #barra 17
                   [nos[14,:],nos[5,:]],                                        #barra 18
                   [nos[14,:],nos[15,:]],                                       #barra 19
                   [nos[5,:],nos[9,:]],                                         #barra 20
                   [nos[5,:],nos[6,:]],                                         #barra 21
                   [nos[6,:],nos[9,:]],                                         #barra 22
                   [nos[6,:],nos[15,:]],                                        #barra 23
                   [nos[6,:],nos[10,:]],                                        #barra 24
                   [nos[9,:],nos[15,:]],                                        #barra 25
                   [nos[15,:],nos[10,:]],                                       #barra 26
                   [nos[1,:],nos[10,:]],                                        #barra 27
                   [nos[1,:],nos[6,:]]])                                        #barra 28
plotar_estrutura(nos, barras_coord, Apoio_A, Apoio_B, Apoio_C, peso_estrutura(nos, barras, area_barra), iterpontofixo(Carga_C,0.001,200))
```

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
abacate = (4-(2)**(1/2))/14 #medida base triangulo 45°
Round = 5 #Numero de digitos depois da virgula
#Posição dos nós
nos = np.array([[0,0], #nó A (0)
                [1,0], #nó B (1)
                [0.5,0], #nó C (2)
                [abacate,0], #nó 3
                [2*abacate,0], #nó 4
                [1-2*abacate,0], #nó 5
                [1-abacate,0], #nó 6
                [abacate/2,abacate/2], #nó 7
                [3*abacate/2, abacate/2], #nó 8
                [1-3*abacate/2,abacate/2], #nó 9
                [1-abacate/2,abacate/2], #nó 10
                [abacate,abacate], #nó 11
                [2*abacate,abacate], #nó 12
                [0.5, abacate], #nó 13
                [1-2*abacate,abacate], #nó 14
                [1-abacate,abacate]]) #nó 15
#conectividade das barras em coordenadas
barras_coord = np.array([[nos[0,:],nos[3,:]],        #barra 0
                   [nos[0,:],nos[7,:]],              #barra 1
                   [nos[3,:],nos[7,:]],              #barra 2
                   [nos[3,:],nos[11,:]],             #barra 3
                   [nos[3,:],nos[8,:]],              #barra 4
                   [nos[3,:],nos[4,:]],              #barra 5
                   [nos[8,:],nos[11,:]],             #barra 6
                   [nos[8,:],nos[4,:]],              #barra 7
                   [nos[7,:],nos[11,:]],             #barra 8
                   [nos[11,:],nos[12,:]],            #barra 9
                   [nos[12,:],nos[13,:]],            #barra 10
                   [nos[12,:],nos[4,:]],             #barra 11
                   [nos[13,:],nos[4,:]],             #barra 12
                   [nos[13,:],nos[2,:]],             #barra 13
                   [nos[13,:],nos[5,:]],             #barra 14
                   [nos[13,:],nos[14,:]],            #barra 15
                   [nos[2,:],nos[4,:]],              #barra 16
                   [nos[2,:],nos[5,:]],              #barra 17
                   [nos[14,:],nos[5,:]],             #barra 18
                   [nos[14,:],nos[15,:]],            #barra 19
                   [nos[5,:],nos[9,:]],              #barra 20
                   [nos[5,:],nos[6,:]],              #barra 21
                   [nos[6,:],nos[9,:]],              #barra 22
                   [nos[6,:],nos[15,:]],             #barra 23
                   [nos[6,:],nos[10,:]],             #barra 24
                   [nos[9,:],nos[15,:]],             #barra 25
                   [nos[15,:],nos[10,:]],            #barra 26
                   [nos[1,:],nos[10,:]],             #barra 27
                   [nos[1,:],nos[6,:]]])             #barra 28
#conectividade das barras em inteiros
barras = np.array([[0,3],              #barra 0
                   [0,7],              #barra 1
                   [3,7],              #barra 2
                   [3,11],             #barra 3
                   [3,8],              #barra 4
                   [3,4],              #barra 5
                   [8,11],             #barra 6
                   [8,4],              #barra 7
                   [7,11],             #barra 8
                   [11,12],            #barra 9
                   [12,13],            #barra 10
                   [12,4],             #barra 11
                   [13,4],             #barra 12
                   [13,2],             #barra 13
                   [13,5],             #barra 14
                   [13,14],            #barra 15
                   [2,4],              #barra 16
                   [2,5],              #barra 17
                   [14,5],             #barra 18
                   [14,15],            #barra 19
                   [5,9],              #barra 20
                   [5,6],              #barra 21
                   [6,9],              #barra 22
                   [6,15],             #barra 23
                   [6,10],             #barra 24
                   [9,15],             #barra 25
                   [15,10],            #barra 26
                   [1,10],             #barra 27
                   [1,6]])             #barra 28
#areas barras
area_barra = np.array([[1.E-4],              #área da barra 0
                       [1.E-4],              #área da barra 1
                       [1.E-4],              #área da barra 2
                       [1.E-4],              #área da barra 3
                       [1.E-4],              #área da barra 4
                       [1.E-4],              #área da barra 5
                       [1.E-4],              #área da barra 6
                       [1.E-4],              #área da barra 7
                       [1.E-4],              #área da barra 8
                       [1.E-4],              #área da barra 9
                       [1.E-4],              #área da barra 10
                       [1.E-4],              #área da barra 11
                       [1.E-4],              #área da barra 12
                       [1.E-4],              #área da barra 13
                       [1.E-4],              #área da barra 14
                       [1.E-4],              #área da barra 15
                       [1.E-4],              #área da barra 16
                       [1.E-4],              #área da barra 17
                       [1.E-4],              #área da barra 18
                       [1.E-4],              #área da barra 19
                       [1.E-4],              #área da barra 20
                       [1.E-4],              #área da barra 21
                       [1.E-4],              #área da barra 22
                       [1.E-4],              #área da barra 23
                       [1.E-4],              #área da barra 24
                       [1.E-4],              #área da barra 25
                       [1.E-4],              #área da barra 26
                       [1.E-4],              #área da barra 27
                       [1.E-4]])             #área da barra 28
#Numero dos nos A B
Apoio_A = 0
Apoio_B = 1      
#Numero do no C Carga C
Apoio_C = 2
#Carga Maxima = 167.6
Carga_C = 100  
#Calcular peso da estrutra
def peso_estrutura(Nos, Barras, Areas):
  Densidade = 1.547*10**3
  Peso = 0
  for i in range(0,np.shape(Barras[:,0,0])[0]):                  #Somar o peso de todas as barras
    Peso = Densidade * Areas[i,0] * ((Barras[i,0,0]-Barras[i,1,0])**2+(Barras[i,0,1]-Barras[i,1,1])**2)**(1/2) + Peso
  return(Peso)
Peso = peso_estrutura(nos, barras_coord, area_barra)
num_nos = np.shape(nos)[0]
num_barras = np.shape(barras)[0]
E = 3.6*10**8
from math import *
def solucao_estrutural(nodes,barras,area_bar,E,nnodes,nbarras,
					   no_apoio_A,no_apoio_B,no_carga,carga_aplicada):

	# ________________________________________
	#
	# ALGORITMO DE SOLUÇÃO ESTRUTURAL DE
	# TRELIÇA PLANA LINEAR
	# ________________________________________
	#
	# > nodes          : coordenadas x e y de cada nó
	# > barras         : nó inicial e nó final de cada barra
	# > area_bar       : a área da seção transversal de cada barra
	# > E              : módulo de elasticidade do material
	# > nnodes         : número de nós
	# > nbarras        : número de barras
	# > no_apoio_A     : apoio 1 (ux = uy = 0.0)
	# > no_apoio_B     : apoio 2 (ux = uy = 0.0)
	# > no_carga       : nó onde a carga F = (0.0,carga_aplicada) é aplicada
	# > carga_aplicada : valor da carga vertical aplicada 
	# < fbarra         : força na direção de cada barra ( tração = (+), compressão = (-) )
	# ________________________________________

#copia dados da geometria da estrutura	
	nodesc = np.copy(nodes)
	barrasc = np.copy(barras)
	area_barc = np.copy(area_bar)
#inicializa arranjos
	l_bar = np.zeros( barrasc.shape[0] )
	cθ_bar = np.zeros( barrasc.shape[0] )
	sθ_bar = np.zeros( barrasc.shape[0] )
	kl = np.zeros( (4,4,nbarras) )
	Kg = np.zeros( (2*nnodes,2*nnodes) )
	Fg = np.zeros( (2*nnodes,1) )
	cc = np.zeros( (2*nnodes,1),dtype='int32' )
	fbarra = np.zeros (nbarras)
#aplica a carga (carga_aplicada) na direção y negativa do nó no_carga
	Fg[2*no_carga+1,0] = -carga_aplicada
#aplica condição de contorno de apoio (deslocamentos nulos) no nós no_apoio_A e no_apoio_B
	cc[2*no_apoio_A,0]   = 1
	cc[2*no_apoio_A+1,0] = 1
	cc[2*no_apoio_B,0]   = 1
	cc[2*no_apoio_B+1,0] = 1
#calcula comprimendo e ângulos das barras
	for i in range(0,nbarras):
		no1 = barrasc[i,0]
		no2 = barrasc[i,1]
		dx = nodesc[no1,0]-nodesc[no2,0]
		dy = nodesc[no1,1]-nodesc[no2,1]
		l_bar[i] = sqrt(dx**2 + dy**2)
		cθ_bar[i] = -dx/l_bar[i]
		sθ_bar[i] = -dy/l_bar[i]
#calcula rigidez kl de cada barra:
	for i in range(0,nbarras):
		kb = ( E * area_barc[i] / l_bar[i] )
		kl[0,0,i] =  kb * ( cθ_bar[i] )**2
		kl[0,1,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[0,2,i] = -kb * ( cθ_bar[i] )**2
		kl[0,3,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[1,0,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[1,1,i] =  kb * ( sθ_bar[i] )**2
		kl[1,2,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[1,3,i] = -kb * ( sθ_bar[i] )**2
		kl[2,0,i] = -kb * ( cθ_bar[i] )**2
		kl[2,1,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[2,2,i] =  kb * ( cθ_bar[i] )**2
		kl[2,3,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[3,0,i] = -kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[3,1,i] = -kb * ( sθ_bar[i] )**2
		kl[3,2,i] =  kb * ( cθ_bar[i] ) * ( sθ_bar[i] )
		kl[3,3,i] =  kb * ( sθ_bar[i] )**2
#montagem da rigidez Kg do sistema linear:
	for i in range(0,nbarras):
		no1 = barrasc[i,0]
		no2 = barrasc[i,1]
		noAx = no1*2
		noAy = noAx + 1
		noBx = no2*2
		noBy = noBx + 1
		Kg[noAx,noAx] = Kg[noAx,noAx] + kl[0,0,i]
		Kg[noAx,noAy] = Kg[noAx,noAy] + kl[0,1,i]
		Kg[noAy,noAx] = Kg[noAy,noAx] + kl[1,0,i]
		Kg[noAy,noAy] = Kg[noAy,noAy] + kl[1,1,i]
		Kg[noAx,noBx] = Kg[noAx,noBx] + kl[0,2,i]
		Kg[noAx,noBy] = Kg[noAx,noBy] + kl[0,3,i]
		Kg[noAy,noBx] = Kg[noAy,noBx] + kl[1,2,i]
		Kg[noAy,noBy] = Kg[noAy,noBy] + kl[1,3,i]
		Kg[noBx,noAx] = Kg[noBx,noAx] + kl[2,0,i]
		Kg[noBy,noAx] = Kg[noBy,noAx] + kl[2,1,i]
		Kg[noBx,noAy] = Kg[noBx,noAy] + kl[3,0,i]
		Kg[noBy,noAy] = Kg[noBy,noAy] + kl[3,1,i]
		Kg[noBx,noBx] = Kg[noBx,noBx] + kl[2,2,i]
		Kg[noBx,noBy] = Kg[noBx,noBy] + kl[2,3,i]
		Kg[noBy,noBx] = Kg[noBy,noBx] + kl[3,2,i]
		Kg[noBy,noBy] = Kg[noBy,noBy] + kl[3,3,i]
#aplica condições de contorno sobre a rigidez Kg do sistema linear	
	for i in range(0,2*nnodes):
		if ( cc[i] == 1 ):
			Kg[:,i] = 0.0
			Kg[i,:] = 0.0
			Kg[i,i] = 1.0
#calcula os deslocamento x e y de cada nó do sistema linear Ax=b,
#onde A=Kg, x=uxy, b=Fg, empregando eliminação de Gauss
	uxy = np.linalg.solve(Kg,Fg)
#calcula força interna na direção de cada barra
	for i in range(0,nbarras):
		no1 = barrasc[i,0]
		no2 = barrasc[i,1]
		ul1x = uxy[no1*2]
		ul1y = uxy[no1*2+1]
		ul2x = uxy[no2*2]
		ul2y = uxy[no2*2+1]
		fbarra[i] = ( ( cθ_bar[i]*kl[2,0,i] + sθ_bar[i]*kl[3,0,i] ) * ul1x + 
    				  ( cθ_bar[i]*kl[2,1,i] + sθ_bar[i]*kl[3,1,i] ) * ul1y +
					  ( cθ_bar[i]*kl[2,2,i] + sθ_bar[i]*kl[3,2,i] ) * ul2x +
					  ( cθ_bar[i]*kl[2,3,i] + sθ_bar[i]*kl[3,3,i] ) * ul2y )
#saída de valores
	return fbarra
forca_barras = solucao_estrutural(nos,barras,area_barra,E,num_nos,num_barras,
				             	  Apoio_A,Apoio_B,Apoio_C,Carga_C)
for i in range(0,np.shape(forca_barras)[0]):
	print('Força na barra entre os nos '+ str(barras[i,0])+'-'+str(barras[i,1])+' / Barra '+ str(i)+' = {:^+15.6E}'.format(float(round(forca_barras[i],5))) )
#Plota a estrutura
def plotar_estrutura(Nos, Barras, N_apoio_A, N_apoio_B, N_apoio_C, Pes, Carga, Round):
  plt.figure(1, figsize=(8,8))
  plt.grid(True)
  plt.title('PESO = '+str(round(Pes, Round))+' Kg')
  plt.xlabel('x', fontsize = 15)
  plt.ylabel('y', fontsize = 15)
  plt.axis([-0.2, +1.2, -0.2, np.max(nos[:,1])+0.2])             #Defini eixos com margem 0.2     
  for i in range(0,np.shape(Barras[:,0,0])[0]):                  #Plota todas as barras
    plt.plot(Barras[i,:,0], Barras[i,:,1], c='b')
  plt.scatter(Nos[:,0], Nos[:,1], c='r')                         #Plota todos os nos
#Plota os apoios AB e seta 
  plt.plot(Nos[N_apoio_B,0],Nos[N_apoio_B,1]-0.01,'^',c='k',ms=15.0)
  plt.plot(Nos[N_apoio_A,0],Nos[N_apoio_A,1]-0.01,'^',c='k',ms=15.0)
  plt.arrow(Nos[N_apoio_C,0],Nos[N_apoio_C,1],0, -0.030, head_width=0.02, head_length=0.020, fc='k', ec='k')
  plt.show()
plotar_estrutura(nos, barras_coord, Apoio_A, Apoio_B, Apoio_C, Peso, Carga_C, Round)
Rimin = 10000000
for i in range(0, num_barras):
  if(forca_barras[i] >=0.001):
    Ri = 1.676*10**6-(forca_barras[i]/area_barra[i,0])
  if(forca_barras[i] <0.001):
    Termomin = (np.pi*E*area_barra[i,0])/(4*(((barras_coord[i,0,0]-barras_coord[i,1,0])**2+(barras_coord[i,0,1]-barras_coord[i,1,1])**2)**(1/2))**2)
    if(Termomin < 1.676*10**6):
      Ri = Termomin - (forca_barras[i]/area_barra[i,0])
    else:
      Ri = 1.676*10**6 - (forca_barras[i]/area_barra[i,0])
  if(Ri<Rimin):
    Rimin = Ri
#Iteracao
def f(x):
  forca_barras = solucao_estrutural(nos,barras,area_barra,E,num_nos,num_barras, Apoio_A,Apoio_B,Apoio_C,x)
  Rimin = 10000000
  for i in range(0, num_barras):
    if(forca_barras[i] >=0.001):
      Ri = 1.676*10**6-(forca_barras[i]/area_barra[i,0])
    if(forca_barras[i] <0.001):
      Termomin = (np.pi*E*area_barra[i,0])/(4*(((barras_coord[i,0,0]-barras_coord[i,1,0])**2+(barras_coord[i,0,1]-barras_coord[i,1,1])**2)**(1/2))**2)
      if(Termomin < 1.676*10**6):
        Ri = Termomin - (forca_barras[i]/area_barra[i,0])
      if(Termomin>=1.676*10**6):
        Ri = 1.676*10**6 - (forca_barras[i]/area_barra[i,0])
    if(Ri<Rimin):
      Rimin = Ri
  return Rimin

def iterpontofixo(x0,es,imax):
# ________________________________________
#
# ALGORITMO DE ITERAÇÃO DE PONTO FIXO
# ________________________________________
#
# > x0: x inicial
# > es: tolerância de erro
# > imax: número máximo de iterações
# > info: True para apresentar dados da tela
# < xr : raíz da função f
# ________________________________________

  for i in range(0,imax+1):
    if(x0>50):
      if ( i == 0 ):
        Rimin1 = f(x0)
        xi = x0 + Rimin1*x0/10/(1.676*10**6) 
      else:
        Rimin1 = f(xi)
        xi = xi +  Rimin1*x0/10/(1.676*10**6)
      er = ((Rimin1)**2)**(1/2)
      if ( er < es ):
        break
    if(x0<50):
      if ( i == 0 ):
        Rimin1 = f(x0)
        xi = x0 + Rimin1*(x0+5)/(1.676*10**6) 
      else:
        Rimin1 = f(xi)
        xi = xi +  Rimin1*(x0+5)/(1.676*10**6)
      er = ((Rimin1)**2)**(1/2)
      if ( er < es ):
        break
  return xi
CI = Carga_C          #Carga inicial
Tolerancia = 0.0001   #Tolerancia
Iteracoes = 200       #Numero de Iteracoes
print('Carga de colapso =', round(iterpontofixo(CI,Tolerancia,Iteracoes), Round), 'kg')
```
